```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.array([19.442,	49.626,	79.59,	99.512,	114.48,	129.453,144.435,	159.356,	174.322,	189.308,	204.268,	229.257,	259.143,	289.103])
y=np.array([0,	0.000299868,	0.0112254,	0.0167685,	0.0306574,	0.0335911,	0.040897,	0.0463762,	0.0534266,	0.0594469	,0.0717717	,0.0999303,	0.149065,	0.20471
])
```

```python
plt.plot(x,y,'o')
```

```python
def f(x,a,b):
    return a*x +b
```

```python
from numpy import arange
from pandas import read_csv
from scipy.optimize import curve_fit
from matplotlib import pyplot
```

```python
popt1,_ = curve_fit(f,x[:10],y[:10])
popt2,_ = curve_fit(f,x[9:],y[9:])
```

```python
erry= np.array([0,0.00398846,	0.00407783	,0.00415027	,0.00418294	,0.0042427	,0.00434384	,0.00437471,	0.00448211,	0.00453275,	0.00470035	,0.00500707,	0.00577199,	0.00712673
])
```

```python
popt1_3,_ = curve_fit(f,x[:10],3*y[:10])
popt2_3,_ = curve_fit(f,x[9:],3*y[9:])
```

```python
p1=np.poly1d(popt1_3)
p1_der=np.polyder(p1)
mean_k1=0.00138/p1_der[0]
p2=np.poly1d(popt2_3)
p2_der=np.polyder(p2)
mean_k2=0.00138/p2_der[0]
print(mean_k1,mean_k2)
```

```python
intercept_x = (popt2[1] -popt1[1]) / (popt1[0] - popt2[0])
intercept_y = popt1[0] * intercept_x + popt1[1]
plt.scatter(intercept_x, intercept_y, color='green', zorder=5)
plt.annotate('Intercept', (intercept_x, intercept_y), textcoords="offset points", xytext=(0,10), ha='center')
print(intercept_x,intercept_y)
```

```python
plt.figure(figsize=(10,6))
plt.errorbar(x,y,yerr=erry,fmt='o',elinewidth=1,capsize=2)
plt.plot(x[:11],f(x[:11],*popt1),'r--',linewidth=1,label='k={:.4f}'.format(mean_k1))
plt.plot(x[9:],f(x[9:],*popt2),'b--',linewidth=1,label='k={:.4f}'.format(mean_k2))
plt.scatter(intercept_x, intercept_y, color='green', zorder=5)
plt.annotate('Intercept', (intercept_x, intercept_y), textcoords="offset points", xytext=(10,-20), ha='center')
plt.xlabel(r'$T~ (K)$',fontsize=14)
plt.ylabel(r'$\langle x^2 \rangle\ (\AA^2)$',fontsize=14)
plt.legend(fontsize=14)
plt.savefig('x_2 vs T',dpi=250)
plt.show()
```
